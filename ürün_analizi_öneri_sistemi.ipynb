#**2-Veri Hazırlama-Veri Ön İşleme**

Veri analizi ve görselleştirmeye dalmadan önce, veri setimizin temiz ve düzgün biçimlendirilmiş olduğundan emin olmamız gerekir. Bu işleme veri hazırlama denir ve birkaç adım içerir:

2.1 🔍 Veri Denetimi: Eksik değerler, yinelenenler veya tutarsız veriler olup olmadığını görmek için veri setini inceleyerek başlayacağız. Ayrıca veri türlerinin doğru olup olmadığını kontrol edeceğiz ve veri setinin analize hazır olduğundan emin olacağız.

2.2 🧹 Veri Temizleme: Daha sonra, hataları, tutarsızlıkları veya alakasız bilgileri kaldırarak veya düzelterek veri setini temizleyeceğiz. Bu, veri setini daha güvenilir ve doğru hale getirecektir.

2.3 📏 Veri Dönüşümü: Veri setini temizledikten sonra, analiz için daha kullanışlı hale getirmek üzere verileri dönüştürmemiz gerekebilir. Bu, ölçekleme, normalleştirme veya özellik mühendisliğini içerebilir.

2.4 💾 Veri Kaydetme: Verileri hazırladıktan sonra, orijinal veri setinin üzerine yazmaktan kaçınmak için yeni bir dosyaya kaydedeceğiz. Bu şekilde, ihtiyaç duyduğumuzda her zaman orijinal veri setine geri dönebiliriz.

Bir veri kümesindeki bazı değişkenler nesne veri türüne sahip olabilir, yani dizelerdir. Bu değişkenler üzerinde sayısal analiz yapmak için bunları sayısal değerlere dönüştürmemiz gerekir. Örneğin, tüm ürünlerin toplam fiyatını hesaplamak istiyorsak, fiyat değişkeni nesne biçimindeyse bunu yapamayız. Önce onu sayısal bir veri türüne dönüştürmemiz gerekir.

```python
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

Veri türlerini ayarladığımıza göre, veritabanımızda bulunması ilginç olabilecek bir sütun daha oluşturalım. **"rating_weighted"** çünkü yalnızca ortalama puanı değil, aynı zamanda ürünü derecelendiren kişi sayısını da dikkate almanın bir yolu olarak oluşturulabilir. Bu sütun, ortalama puanı derecelendirme sayısına göre tartarak, çok sayıda derecelendiricisi olan derecelendirmelere daha fazla ağırlık verir. Bu, yüksek ortalama derecelendirmeleri olan ancak az sayıda derecelendiricisi olan ürünlere kıyasla yüksek müşteri memnuniyeti ve çok sayıda olumlu derecelendirmesi olan ürünleri belirlemeye yardımcı olabilir.

```python
# Creating the column "rating_weighted"
df['rating_weighted'] = df['rating'] * df['rating_count']
```

```python
import re
# Öncelikle modify_url fonksiyonunu tanımlıyoruz
def modify_url(url):
    # Eğer URL zaten doğru formattaysa, olduğu gibi döndür
    if re.match(r"^https://m\.media-amazon\.com/images/I/.+\.jpg$", url):
        return url

    # Sabit başlangıç kısmı
    fixed_part = "https://m.media-amazon.com/images/I/"

    # URL'deki 'images/W/WEBP_' kısmını düzenle
    pattern = re.compile(r"https://m\.media-amazon\.com/images/[^/]+/WEBP_([^/]+)/images/I/([^/]+)\.jpg")
    match = pattern.search(url)

    if match:
        # Elde edilen kısmı sabit başlangıç kısmına ekleyin
        modified_url = fixed_part + match.group(2) + ".jpg"
        return modified_url
    else:
        return "Geçerli URL bulunamadı!"
```

```python
df['img_link'] = df['img_link'].apply(modify_url)
```

# 3-Keşfedici Veri Analizi (EDA) + Veri görselleştirme

3.1 Bir çubuk grafiği kullanarak ürünlerin kategoriye göre dağılımını analiz edin.

3.2 Bir histogram kullanarak müşteri derecelendirmelerinin dağılımını analiz edin.

3.3 İncelemelerde kullanılan en yaygın kelimelerin kelime bulutlarını veya sıklık tablolarını oluşturarak incelemeleri analiz edin.

3.4 Ürün fiyatı ile müşteri derecelendirmesi arasındaki ilişki gibi farklı özellikler arasındaki herhangi bir korelasyonu belirlemek için istatistiksel analiz gerçekleştirin.

```python
import matplotlib.pyplot as plt

#3.1
# Ürünlerin ana kategoriye göre dağılımının analizi

main_category_counts = df['main_category'].value_counts()[:30] # Select only the top 30 main categories.
plt.bar(range(len(main_category_counts)), main_category_counts.values)
plt.ylabel('Ürün Sayısı')
plt.title('Ürünlerin Ana Kategoriye Göre Dağılımı (İlk 30)')
plt.xticks(range(len(main_category_counts)), '') # hide X-axis labels
plt.show()

# En iyi 30 ana kategori
top_main_categories = pd.DataFrame({'Main Category': main_category_counts.index, 'Ürün Sayısı': main_category_counts.values})
print('İlk 30 ana kategori:')
print(top_main_categories.to_string(index=False))
```

🔍 İlk üç ana kategori Elektronik, Bilgisayarlar ve Aksesuarlar ve Ev ve Mutfak'tır. Bu, bu kategorilerin müşteriler arasında popüler olduğunu gösterir.

🔍 Diğer ana kategorilerdeki ürün sayısı oldukça düşüktür ve bu kategorilerin ilk üç kadar popüler olmadığını gösterir.

🔍 Ofis Ürünleri, Müzik Aletleri, Ev Dekorasyonu, Oyuncaklar ve Oyunlar, Araba ve Motosiklet ve Sağlık ve Kişisel Bakım kategorilerinde çok az ürün vardır ve bu, bu kategorilere olan talebin daha az olduğunu gösterebilir.

```python
# Ürünlerin son kategoriye göre dağılımının analizi

sub_category_counts = df['sub_category'].value_counts()[:30] # Select only the

plt.bar(range(len(sub_category_counts)), sub_category_counts.values)
plt.ylabel('Ürün Sayısı')
plt.title('Ürünlerin Alt Kategoriye Göre Dağılımı (İlk 30)')
plt.xticks(range(len(sub_category_counts)), '') # hide X-axis labels
plt.show()


top_sub_categories = pd.DataFrame({'Sub Category': sub_category_counts.index, 'Number of Products': sub_category_counts.values})
print('İlk 30 alt kategori:')
print(top_sub_categories.to_string(index=False))
```

```python
# 3.2 Müşteri puanlarının dağılımını bir histogram kullanarak analiz edin.
# Plot histogram
plt.hist(df['rating'])
plt.xlabel('Derecelendirme')
plt.ylabel('İnceleme Sayısı')
plt.title('Müşteri Puanlarının Dağılımı')
plt.show()
```

# 4. Öneri Model

Verileri kullanarak, bir kullanıcının ilgi alanlarına ve geçmiş satın alımlarına göre ürünler öneren kişiselleştirilmiş bir öneri sistemi oluşturacağım.

Bu, bir kullanıcı oturum açtığında, platformumuzla olan geçmişine göre öneriler alacağı anlamına gelir. 🤩 Kendileriyle ilgili yeni ve heyecan verici ürünler bulabilecekler ve bu da alışveriş deneyimlerini daha keyifli ve rahat hale getirecek.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['user_id_encoded'] = le.fit_transform(df['user_id'])

# user_id frekans tablosuyla yeni bir veri çerçevesi oluşturma

freq_table = pd.DataFrame({'User ID': df['user_id_encoded'].value_counts().index, 'Frequency': df['user_id_encoded'].value_counts().values})

# Display the dataframe
print(freq_table)
id_example = freq_table.iloc[0,0]
print(id_example)
```

Bir dize sütununu sayılara dönüştürmek, veri analizi ve makine öğreniminde yaygın bir adımdır ve çeşitli şekillerde faydalı olabilir. 😃

Birincisi, birçok makine öğrenimi algoritması düzgün çalışması için sayısal giriş verilerine ihtiyaç duyar, bu nedenle verileri dönüştürmek algoritmanın verileri doğru şekilde işleyebilmesini sağlamaya yardımcı olabilir. 👨‍💻

İkincisi, sayısal verilerle bazı durumlarda dizelerden daha kolay çalışılabilir. Örneğin, sayısal veriler üzerinde matematiksel işlemler gerçekleştirebilirsiniz; bu, ortalamalar, toplamlar ve diğer istatistikler gibi şeyleri hesaplamada faydalı olabilir. 📊

Bir öneri sistemi bağlamında, kullanıcı kimliklerini dizelerden sayılara dönüştürmek, kullanıcılar veya öğeler arasındaki benzerlik puanlarının hesaplanmasını basitleştirmeye yardımcı olabilir. Bu, öneri sisteminin verimliliğini ve doğruluğunu artırabilir. 🚀

Genel olarak, dizeleri sayılara dönüştürmek veri analizi ve makine öğreniminde yaygın ve faydalı bir adımdır ve özellikle öneri sistemlerinin performansını iyileştirmede yardımcı olabilir. 🤖

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_products(df, user_id_encoded):
    # Ürün açıklamalarını sayısal özellik vektörlerine dönüştürmek için TfidfVectorizer'ı kullanın
    tfidf = TfidfVectorizer(stop_words='english')
    df['about_product'] = df['about_product'].fillna('')  # NaN değerlerini boş dizeyle doldur
    tfidf_matrix = tfidf.fit_transform(df['about_product'])

    # Kullanıcının satın alma geçmişini alın
    user_history = df[df['user_id_encoded'] == user_id_encoded]

    # Her ürün açıklaması çifti arasındaki benzerliği hesaplamak için cosine_similarity'yi kullanın
    # yalnızca kullanıcının daha önce satın aldığı ürünler için
    indices = user_history.index.tolist()

    if indices:
        # Yalnızca satın alınan ürünlere ait satır ve sütunları içeren yeni bir benzerlik matrisi oluşturun
        cosine_sim_user = cosine_similarity(tfidf_matrix[indices], tfidf_matrix)

        # Ürün endekslerini endeks, ürün adlarını da değerler olarak kullanarak bir pandas Serisi oluşturun
        products = df.iloc[indices]['product_name']
        indices = pd.Series(products.index, index=products)

        # Kullanıcının daha önce satın aldığı ürünlere benzer ürünlerin endekslerini ve benzerlik puanlarını alın
        similarity_scores = list(enumerate(cosine_sim_user[-1]))
        similarity_scores = [(i, score) for (i, score) in similarity_scores if i not in indices]

        # Benzerlik puanlarını azalan düzende sıralayın
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

        # En benzer ilk 5 ürünün endekslerini alın
        top_products = [i[0] for i in similarity_scores[1:6]]

        # En çok benzeyen ilk 5 ürünün adını alın
        recommended_products = df.iloc[top_products]['product_name'].tolist()

        # Önerinin skorlarını öğrenme
        score = [similarity_scores[i][1] for i in range(5)]

        # Sonuçlarla bir DataFrame oluşturun
        results_df = pd.DataFrame({'Id Encoded': [user_id_encoded] * 5,
                                   'Önerilen Ğrün': recommended_products,
                                   'Puan Önerisi': score})

        return results_df

    else:
        print("Satın alma geçmişi bulunamadı.")
        return None
```
