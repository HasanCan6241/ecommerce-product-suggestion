```python
import re
# √ñncelikle modify_url fonksiyonunu tanƒ±mlƒ±yoruz
def modify_url(url):
    # Eƒüer URL zaten doƒüru formattaysa, olduƒüu gibi d√∂nd√ºr
    if re.match(r"^https://m\.media-amazon\.com/images/I/.+\.jpg$", url):
        return url

    # Sabit ba≈ülangƒ±√ß kƒ±smƒ±
    fixed_part = "https://m.media-amazon.com/images/I/"

    # URL'deki 'images/W/WEBP_' kƒ±smƒ±nƒ± d√ºzenle
    pattern = re.compile(r"https://m\.media-amazon\.com/images/[^/]+/WEBP_([^/]+)/images/I/([^/]+)\.jpg")
    match = pattern.search(url)

    if match:
        # Elde edilen kƒ±smƒ± sabit ba≈ülangƒ±√ß kƒ±smƒ±na ekleyin
        modified_url = fixed_part + match.group(2) + ".jpg"
        return modified_url
    else:
        return "Ge√ßerli URL bulunamadƒ±!"
```

```python
df['img_link'] = df['img_link'].apply(modify_url)
```

# 4. √ñneri Model

Verileri kullanarak, bir kullanƒ±cƒ±nƒ±n ilgi alanlarƒ±na ve ge√ßmi≈ü satƒ±n alƒ±mlarƒ±na g√∂re √ºr√ºnler √∂neren ki≈üiselle≈ütirilmi≈ü bir √∂neri sistemi olu≈üturacaƒüƒ±m.

Bu, bir kullanƒ±cƒ± oturum a√ßtƒ±ƒüƒ±nda, platformumuzla olan ge√ßmi≈üine g√∂re √∂neriler alacaƒüƒ± anlamƒ±na gelir. ü§© Kendileriyle ilgili yeni ve heyecan verici √ºr√ºnler bulabilecekler ve bu da alƒ±≈üveri≈ü deneyimlerini daha keyifli ve rahat hale getirecek.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['user_id_encoded'] = le.fit_transform(df['user_id'])

# user_id frekans tablosuyla yeni bir veri √ßer√ßevesi olu≈üturma

freq_table = pd.DataFrame({'User ID': df['user_id_encoded'].value_counts().index, 'Frequency': df['user_id_encoded'].value_counts().values})

# Display the dataframe
print(freq_table)
id_example = freq_table.iloc[0,0]
print(id_example)
```

Bir dize s√ºtununu sayƒ±lara d√∂n√º≈üt√ºrmek, veri analizi ve makine √∂ƒüreniminde yaygƒ±n bir adƒ±mdƒ±r ve √ße≈üitli ≈üekillerde faydalƒ± olabilir. üòÉ

Birincisi, bir√ßok makine √∂ƒürenimi algoritmasƒ± d√ºzg√ºn √ßalƒ±≈ümasƒ± i√ßin sayƒ±sal giri≈ü verilerine ihtiya√ß duyar, bu nedenle verileri d√∂n√º≈üt√ºrmek algoritmanƒ±n verileri doƒüru ≈üekilde i≈üleyebilmesini saƒülamaya yardƒ±mcƒ± olabilir. üë®‚Äçüíª

ƒ∞kincisi, sayƒ±sal verilerle bazƒ± durumlarda dizelerden daha kolay √ßalƒ±≈üƒ±labilir. √ñrneƒüin, sayƒ±sal veriler √ºzerinde matematiksel i≈ülemler ger√ßekle≈ütirebilirsiniz; bu, ortalamalar, toplamlar ve diƒüer istatistikler gibi ≈üeyleri hesaplamada faydalƒ± olabilir. üìä

Bir √∂neri sistemi baƒülamƒ±nda, kullanƒ±cƒ± kimliklerini dizelerden sayƒ±lara d√∂n√º≈üt√ºrmek, kullanƒ±cƒ±lar veya √∂ƒüeler arasƒ±ndaki benzerlik puanlarƒ±nƒ±n hesaplanmasƒ±nƒ± basitle≈ütirmeye yardƒ±mcƒ± olabilir. Bu, √∂neri sisteminin verimliliƒüini ve doƒüruluƒüunu artƒ±rabilir. üöÄ

Genel olarak, dizeleri sayƒ±lara d√∂n√º≈üt√ºrmek veri analizi ve makine √∂ƒüreniminde yaygƒ±n ve faydalƒ± bir adƒ±mdƒ±r ve √∂zellikle √∂neri sistemlerinin performansƒ±nƒ± iyile≈ütirmede yardƒ±mcƒ± olabilir. ü§ñ

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def recommend_products(df, user_id_encoded):
    # √úr√ºn a√ßƒ±klamalarƒ±nƒ± sayƒ±sal √∂zellik vekt√∂rlerine d√∂n√º≈üt√ºrmek i√ßin TfidfVectorizer'ƒ± kullanƒ±n
    tfidf = TfidfVectorizer(stop_words='english')
    df['about_product'] = df['about_product'].fillna('')  # NaN deƒüerlerini bo≈ü dizeyle doldur
    tfidf_matrix = tfidf.fit_transform(df['about_product'])

    # Kullanƒ±cƒ±nƒ±n satƒ±n alma ge√ßmi≈üini alƒ±n
    user_history = df[df['user_id_encoded'] == user_id_encoded]

    # Her √ºr√ºn a√ßƒ±klamasƒ± √ßifti arasƒ±ndaki benzerliƒüi hesaplamak i√ßin cosine_similarity'yi kullanƒ±n
    # yalnƒ±zca kullanƒ±cƒ±nƒ±n daha √∂nce satƒ±n aldƒ±ƒüƒ± √ºr√ºnler i√ßin
    indices = user_history.index.tolist()

    if indices:
        # Yalnƒ±zca satƒ±n alƒ±nan √ºr√ºnlere ait satƒ±r ve s√ºtunlarƒ± i√ßeren yeni bir benzerlik matrisi olu≈üturun
        cosine_sim_user = cosine_similarity(tfidf_matrix[indices], tfidf_matrix)

        # √úr√ºn endekslerini endeks, √ºr√ºn adlarƒ±nƒ± da deƒüerler olarak kullanarak bir pandas Serisi olu≈üturun
        products = df.iloc[indices]['product_name']
        indices = pd.Series(products.index, index=products)

        # Kullanƒ±cƒ±nƒ±n daha √∂nce satƒ±n aldƒ±ƒüƒ± √ºr√ºnlere benzer √ºr√ºnlerin endekslerini ve benzerlik puanlarƒ±nƒ± alƒ±n
        similarity_scores = list(enumerate(cosine_sim_user[-1]))
        similarity_scores = [(i, score) for (i, score) in similarity_scores if i not in indices]

        # Benzerlik puanlarƒ±nƒ± azalan d√ºzende sƒ±ralayƒ±n
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

        # En benzer ilk 5 √ºr√ºn√ºn endekslerini alƒ±n
        top_products = [i[0] for i in similarity_scores[1:6]]

        # En √ßok benzeyen ilk 5 √ºr√ºn√ºn adƒ±nƒ± alƒ±n
        recommended_products = df.iloc[top_products]['product_name'].tolist()

        # √ñnerinin skorlarƒ±nƒ± √∂ƒürenme
        score = [similarity_scores[i][1] for i in range(5)]

        # Sonu√ßlarla bir DataFrame olu≈üturun
        results_df = pd.DataFrame({'Id Encoded': [user_id_encoded] * 5,
                                   '√ñnerilen ƒûr√ºn': recommended_products,
                                   'Puan √ñnerisi': score})

        return results_df

    else:
        print("Satƒ±n alma ge√ßmi≈üi bulunamadƒ±.")
        return None
```

≈ûimdi, sistemimizden √ßƒ±karƒ±labilecek bazƒ± i√ßg√∂r√ºlere ve bunlarƒ±n bir ≈üirket i√ßin nasƒ±l deƒüerli olabileceƒüine bir g√∂z atalƒ±m:
